```python
black_pais = np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(black_pais)
```

```python
cv.rectangle(black_pais,pt1=(384,0),pt2=(510,150),color=(0,255,0),thickness=5)
plt.imshow(black_pais)
```

```python
cv.rectangle(black_pais,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=10)
plt.imshow(black_pais)
```

```python
cv.circle(black_pais,center=(100,100),radius=50,color=(255,0,0),thickness=5)
plt.imshow(black_pais)
```

```python
cv.line(black_pais,pt1=(0,0),pt2=(512,512),color=(255,255,255),thickness=5)
plt.imshow(black_pais)
```

```python
font=cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_pais,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,0),thickness=5,lineType=cv.LINE_AA)
plt.imshow(black_pais)
```

```python
black_pais2=np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(black_pais2)
```

```python
vertices=np.array([[100,300],
                   [200,200],
                   [400,300],
                   [200,400]],dtype=np.int32)
vertices
```

```python
pts=vertices.reshape((-1,1,2))#fungsi reshape adalah mengubah bentuk array tanpa mengubah data
pts#dimana pada kasus ini data data akan dibentuk menjadi per kolom menambahkan kurung siku
```

```python
cv.polylines(black_pais2,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(black_pais2)
```
